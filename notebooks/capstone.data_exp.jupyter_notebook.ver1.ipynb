* Only 0.1% of total values are missing, coming from 1.3% of missing values in `resort_type_code` variable.
* The report finds numeric, categorical, boolean and text values. However from the Data Dictionary we can see that some columns have been attributed the incorrect data types (e.g. date columns).
* The variables most correlated with the target (`amount_spend_per_room`) are:
    1. `total_pax`

```python
date_feats = ['checkout_date', 'checkin_date', 'booking_date']

category_feats = df.columns[df.dtypes=='object'].tolist() + \
['room_type_booked_code', 'booking_type_code', 'state_code_resort', 'channel_code', 
 'main_product_code', 'persontravellingid', 'resort_region_code', 'resort_type_code']

category_feats = [x for x in category_feats if x not in date_feats]
```

```python
dtypes_dict = {
    "date_feats": date_feats, 
    "category_feats": category_feats}
```

```python
with open('dtypes_dict.pickle', 'wb') as pkl:
    pickle.dump(dtypes_dict, pkl, protocol=pickle.HIGHEST_PROTOCOL)
```
